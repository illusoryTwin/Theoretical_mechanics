#Experiment 1.
**Initial conditions:** $\phi = 15^\circ$, $x = 0$

```python
import csv
import matplotlib.pyplot as plt
import numpy as np


file_path = 'data_test_init1.csv'

test1_positions = []
test1_angles = []
test1_time = []
test2_positions = []
test2_angles = []
test2_time = []
test3_positions = []
test3_angles = []
test3_time = []

# Read data from the CSV file
with open(file_path, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        if reader.line_num == 1:
          continue


        # Reading data for test 1
        # Exctract positions
        test1_positions.append(float(row[1])/1000)
        # Extract angles
        test1_angles.append((float(row[2]))*np.pi/180)
        # Extract time
        time_components = row[0].split(':')
        hours = float(time_components[0])
        minutes = float(time_components[1])
        seconds_millisec = float(time_components[2])
        seconds = int(seconds_millisec)
        milliseconds = (seconds_millisec - seconds) * 1000
        test1_time_val = hours * 3600 * 1000 + minutes * 60 * 1000 + seconds * 1000 + milliseconds
        test1_time.append(test1_time_val)


        # Reading data for test 2
        # Exctract positions
        test2_positions.append(float(row[6])/1000)
        # Extract angles
        test2_angles.append((float(row[7]))*np.pi/180)
        # Extract time
        time_components_2 = row[5].split(':')
        hours_2 = float(time_components_2[0])
        minutes_2 = float(time_components_2[1])
        seconds_millisec_2 = float(time_components_2[2])
        seconds_2 = int(seconds_millisec_2)
        milliseconds_2 = (seconds_millisec_2 - seconds_2) * 1000
        test2_time_val = hours * 3600 * 1000 + minutes_2 * 60 * 1000 + seconds_2 * 1000 + milliseconds_2
        test2_time.append(test2_time_val)

        # Reading data for test 3
        # Exctract positions
        test3_positions.append(float(row[11])/1000)
        # Extract angles
        test3_angles.append((float(row[12]))*np.pi/180)
        # Extract time
        time_components_3 = row[10].split(':')
        hours_3 = float(time_components_3[0])
        minutes_3 = float(time_components_3[1])
        seconds_millisec_3 = float(time_components_3[2])
        seconds_3 = int(seconds_millisec_3)
        milliseconds_3 = (seconds_millisec_3 - seconds_3) * 1000
        test3_time_val = hours_3 * 3600 * 1000 + minutes_3 * 60 * 1000 + seconds_3 * 1000 + milliseconds_3
        test3_time.append(test3_time_val)

first_element = test1_time[0]
test1_time = [time - first_element for time in test1_time]
```

```python
# Calculate the mean position values based on three experiments

positions_array = np.array([test1_positions, test2_positions, test3_positions])
angles_array = np.array([test1_angles, test2_angles, test3_angles])

# Calculate the mean along the rows (axis=0)
mean_positions_test1 = np.mean(positions_array, axis=0)
mean_angles_test1 = np.mean(angles_array, axis=0)

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
plt.title('Initial condition 1, φ = 15°, x = 0')

ax1.plot(test1_time, mean_positions_test1, label='Mean Positions')
ax1.set_ylabel('Position')

ax2.plot(test1_time, mean_angles_test1, label='Mean Angles')
ax2.set_ylabel('Angle')

plt.xlabel('Time')
ax1.legend()
ax2.legend()

plt.show()
```

#Analytical solution

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

T, N, m1, m2, g, k, a2, phi, x, l, J, t, k_air = sp.symbols('T N m1 m2 g k a2 phi x l J t k_air ')

# Define x(t) and its derivatives
x = sp.Function('x')(t)
dx_dt = sp.diff(x, t)
ddx_ddt_sym = sp.diff(x, t, t)

# Define phi(t) and its derivatives
phi = sp.Function('phi')(t)
dphi_dt = sp.diff(phi, t)
ddphi_ddt_sym = sp.diff(phi, t, t)

# Define equations
# eq1 = -T*sp.sin(phi) - m1 * ddx_ddt_sym + k*m1*g*sp.tanh(1000*dx_dt)
# eq2 = N - m1 * g - T*sp.cos(phi)
# eq3 = T*sp.sin(phi) - m2 * (ddphi_ddt_sym * l * sp.cos(phi) - dphi_dt**2 * l * sp.sin(phi)) - m2 * ddx_ddt_sym - k_air*sp.cos(phi)
# eq4 = T*sp.cos(phi) - m2 * g - m2 * (ddphi_ddt_sym * l * sp.sin(phi) + dphi_dt**2 * l * sp.cos(phi)) - k_air*sp.sin(phi)* sp.tanh(1000*dx_dt)

# # Solve the equations
# solutions = sp.solve((eq1, eq3, eq5), (T_x, ddx_ddt_sym, ddphi_ddt_sym))


eq1 = -T*sp.sin(phi) + m1 * ddx_ddt_sym + k*m1*g*sp.tanh(1000*dx_dt)
eq2 = N - m1 * g - T*sp.cos(phi)
eq3 = T*sp.sin(phi) + m2 * (ddphi_ddt_sym * l * sp.cos(phi) - dphi_dt**2 * l * sp.sin(phi) + ddx_ddt_sym) + k_air*sp.sin(phi)*sp.tanh(1000*dx_dt)
eq4 = T*sp.cos(phi) - m2 * g - m2 * (ddphi_ddt_sym * l * sp.sin(phi) + dphi_dt**2 * l * sp.cos(phi)) + k_air*sp.cos(phi)* sp.tanh(1000*dx_dt)

solution = sp.solve((eq1, eq2, eq3, eq4), (T, N, ddx_ddt_sym, ddphi_ddt_sym))



substitution = {m1: 5, m2: 1, l: 1, g: 9.807, k: 0.2, k_air: 3}

# # Substite J = m2 * l**2
numeric_solutions = {key: value.subs(substitution) for key, value in solution.items()}

print(numeric_solutions)
def state_space_model(state, t_val):
    phi_val, x_val, dphi_dt_val, dx_dt_val = state
    ddphi_ddt_val = numeric_solutions[ddphi_ddt_sym].subs(dx_dt, dx_dt_val).subs(dphi_dt, dphi_dt_val).subs(phi, phi_val).evalf().doit()
    ddx_ddt_val = numeric_solutions[ddx_ddt_sym].subs(dx_dt, dx_dt_val).subs(dphi_dt, dphi_dt_val).subs(phi, phi_val).evalf().doit()
    return [dphi_dt_val, dx_dt_val, ddphi_ddt_val, ddx_ddt_val]


# Initial conditions
initial_states = [
    [15.0/180*np.pi, 0.0, 0.0, 0.0],
    [45.0/180*np.pi, 0.25, 0.0, 0.0],
    [-135.0/180*np.pi, 0.25, 0.0, 0.0]
]

ts = np.linspace(0, 10, 100)

# # Plotting
fig, axs = plt.subplots(len(initial_states), 4, figsize=(22, 11))

for i, initial_state in enumerate(initial_states):
    solution = odeint(state_space_model, initial_state, ts)
    phi_result, x_result, phi_dot_result, x_dot_result = solution[:, 0], solution[:, 1], solution[:, 2], solution[:, 3]
    axs[i, 0].plot(ts, x_result, label='X')
    axs[i, 0].set_title(f'Initial Conditions {i+1}, X')
    axs[i, 0].legend()
    axs[i, 0].grid(True)
    axs[i, 1].plot(ts, phi_result, label='Phi')
    axs[i, 1].set_title(f'Initial Conditions {i+1}, Phi')
    axs[i, 1].legend()
    axs[i, 1].grid(True)
    axs[i, 2].plot(ts, x_dot_result, label='X Dot')
    axs[i, 2].set_title(f'Initial Conditions {i+1}, X Dot')
    axs[i, 2].legend()
    axs[i, 2].grid(True)
    axs[i, 3].plot(ts, phi_dot_result, label='Phi Dot')
    axs[i, 3].set_title(f'Initial Conditions {i+1}, Phi Dot')
    axs[i, 3].legend()
    axs[i, 3].grid(True)
plt.tight_layout()
plt.show()
```

#Discussion
From the plots, we observe discrepancies in the positions of x, while φ is predicted quite accurately. Although we made assumptions and theoretical suggestions concerning the coefficients of friction and air resistance, their precise values are unknown to us. For more accurate results, determining these values is necessary. Additionally, in our experiment, we neglected the mass of the rod due to its relatively small value. However, for future, more accurate research, considering this mass might be reasonable.

Also, the issue with x might arise from high friction in the bearings, causing the cart to deviate slightly from its initial position. Determining this friction coefficient in practice is a non-trivial task, which could be a subject for further research.
